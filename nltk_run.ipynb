```python
import Frequency
import matplotlib.pyplot as plt
import nltk
import numpy as np
import re
import sys

from nltk.corpus import stopwords
from wordcloud import WordCloud
```

```python
path = "..\dataset\The Old Man and The Sea.txt"

text = Frequency.read_file(path) # 读取文件
corpus = Frequency.generate_corpus(text)# 文本分词
filtered_words = Frequency.generate_filtered_words(corpus)# 去除停用词
filtered_words = Frequency.remove_suffix_prefix(filtered_words)# 去除前后缀
filtered_words = Frequency.generate_filtered_words(filtered_words)# 二次过滤
tagged_words = Frequency.word_tagging(filtered_words)# 词性标注
lemmas = Frequency.lemmatized_words(tagged_words)# 词形还原
freq_dist = nltk.FreqDist(lemmas)# 词频统计
```

```python
Frequency.WordCloudGeneration('old_man_and_sea', freq_dist)# 词云生成
```
